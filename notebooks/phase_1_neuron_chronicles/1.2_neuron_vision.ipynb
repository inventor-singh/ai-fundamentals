## üß† Section 1: Visualizing the Neuron's Decision Boundary

A single neuron in 2D draws a straight line to separate classes. Let's see this in action!

````python
# Generate synthetic 2D data (two classes)
np.random.seed(42)
num_points = 100
X_class0 = np.random.randn(num_points, 2) - 1.5
X_class1 = np.random.randn(num_points, 2) + 1.5
X = np.vstack([X_class0, X_class1])
y = np.array([0]*num_points + [1]*num_points)

plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', edgecolors='k', s=60, alpha=0.7)
plt.title('Sample Data: Two Classes')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True, alpha=0.3)
plt.show()
````

## üî¨ Section 2: The Neuron's Formula in 2D

A neuron's output:

```
output = sigmoid(w1 * x1 + w2 * x2 + b)
```

The decision boundary is where output = 0.5 (i.e., the neuron is "undecided").
Let's plot this boundary for different weights and bias.

````python
def plot_decision_boundary(w1, w2, b):
    # Plot data
    plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', edgecolors='k', s=60, alpha=0.7)
    
    # Plot decision boundary: w1*x1 + w2*x2 + b = 0
    x_vals = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100)
    if w2 != 0:
        y_vals = -(w1 * x_vals + b) / w2
        plt.plot(x_vals, y_vals, 'k--', linewidth=2, label='Decision Boundary')
    else:
        plt.axvline(-b/w1, color='k', linestyle='--', linewidth=2, label='Decision Boundary')
    
    plt.title(f"Neuron Decision Boundary\n(w1={w1:.2f}, w2={w2:.2f}, b={b:.2f})")
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Example: Try w1=1, w2=1, b=0
plot_decision_boundary(1, 3, 0)
````

## üïπÔ∏è Section 3: Interactive Neuron Explorer

Use the sliders to change the neuron's weights and bias. Watch how the decision boundary moves!

````python
w1_slider = widgets.FloatSlider(value=1, min=-5, max=5, step=0.1, description='w1')
w2_slider = widgets.FloatSlider(value=1, min=-5, max=5, step=0.1, description='w2')
b_slider = widgets.FloatSlider(value=0, min=-5, max=5, step=0.1, description='bias')

ui = widgets.HBox([w1_slider, w2_slider, b_slider])

def interactive_plot(w1, w2, b):
    plot_decision_boundary(w1, w2, b)

out = widgets.interactive_output(interactive_plot, {'w1': w1_slider, 'w2': w2_slider, 'b': b_slider})
display(ui, out)
````

- Changing weights and bias rotates and shifts this line.
